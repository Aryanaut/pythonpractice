```python
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits import mplot3d
```

```python
fig, ax = plt.subplots()
a = 0
loops = int(input())
radius = int(input())
center = input("Enter coords with a space: ")
center = center.split()
coords = []
for coord in center:
    coords.append(int(coord))
xCoords = []
yCoords = []
while a < 360*loops:
    ra = math.radians(a)
    y = (radius*(math.sin(ra)))+coords[1]
    x = (radius*(math.cos(ra)))+coords[0]
    yCoords.append(y)
    xCoords.append(x)
    a += 2
    radius += 1
    
plt.plot(xCoords, yCoords)
plt.scatter(coords[0], coords[1])
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
xCoords = [0, 1, 2]
yCoords = [0, 1, 2]
zCoords = [0, 1, 2]
a = 0
loops = int(input())
radius = int(input())
center = input("Enter coords with a space: ")
center = center.split()
coords = []
for coord in center:
    coords.append(int(coord))
xCoords = []
yCoords = []
zCoords = []
zc = coords[2]
while a < 360*loops:
    ra = math.radians(a)
    y = (radius*(math.sin(ra)))+coords[1]
    x = (radius*(math.cos(ra)))+coords[0]
    zc += 1
    zCoords.append(zc)
    yCoords.append(y)
    xCoords.append(x)
    a += 2
    radius += 0.1
    
plt.plot(xCoords, yCoords, zCoords)
plt.scatter(coords[0], coords[1], coords[2])
```
